# Grafos com NetworkX

Um breve tutorial.

---

## Importação dos pacotes

```python
import networkx as nx
```

## Criação e desenho de um grafo simples

```python
G = nx.Graph()

G.add_edge(1, 2)
G.add_edge(1, 4)
G.add_edge(3, 4)
G.add_edge(4, 5)
G.add_edge(2, 5)

G.add_edge('A', 'B', weight=2)
G.add_edge('B', 'C', weight=4)

nx.draw_spring(G, with_labels=True)
```

## Criação e desenho de um grafo dirigido e sua matriz de adjacências

```python
edges = [(0,1), (1,2), (0,3), (1,3), (3,4)]

G = nx.DiGraph()
G.add_edges_from(edges)

print(nx.adjacency_matrix(G))
nx.draw_circular(G, with_labels=True)
```

## Apresentando algumas métricas

```python
edges = [(0,1), (1,2), (0,3), (1,3), (3,4), (4,5)]

G = nx.Graph()
G.add_edges_from(edges)

print('Degree:', G.degree)
print('Density:', nx.density(G))
print('Diameter:', nx.diameter(G))

print('\nCENTRALITY:\n-----------')
print('Degree:', nx.degree_centrality(G))
print('Closeness:', nx.closeness_centrality(G))
print('Eigenvector:', nx.eigenvector_centrality(G))
print('Betweenness:', nx.betweenness_centrality(G))

nx.draw_circular(G, with_labels=True)
```

## Componentes conexos

```python
E1 = [(0,1), (1,2), (0,3), (1,3), (3,4), (4,5), (2,5)]
E2 = [(6, 7), (6, 8), (8, 9), (8, 10), (6, 9)]
E3 = [(11, 12), (12, 13), (12, 14), (13, 15), (13, 16), (14, 17), (16, 17)]

G = nx.Graph()
G.add_edges_from(E1)
G.add_edges_from(E2)
G.add_edges_from(E3)

print(list(nx.connected_components(G)))

nx.draw(G, with_labels=True)
```

## Caminhos mínimos

```python
edges = [(0,1,3), (1,2,3), (0,3,1), (0,4,5), (1,3,6), (3,4,8), (4,5,3), (3,6,5), (6,7,1), (4,7,4)]
G = nx.DiGraph()
G.add_weighted_edges_from(edges)

pos=nx.planar_layout(G)
nx.draw_networkx(G, pos)
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=nx.get_edge_attributes(G, 'weight')
)

print(nx.dijkstra_path(G, 0, 7))
```
